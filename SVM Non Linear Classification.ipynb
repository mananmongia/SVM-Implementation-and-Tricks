```python
from sklearn.datasets import make_circles
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Make Dataset

```python
X , Y = make_circles(n_samples=500, noise=0.025)
```

```python
print(X.shape, Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c = Y)
```

# Maping to higher dimension

```python
def phi(X):
    """Non Linear Transformation"""
    
    X_ = np.zeros((500,3))
    
    X_[:,:-1] = X
    X_[:,-1] = X[:,0]**2 + X[:,1]**2
    
    return X_
```

# Plots

```python
def plt3D(X,show = True):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection = '3d')
    
    ax.scatter(X[:,0],X[:,1],X[:,2],s = 8 ,c = Y,zdir = 'z', depthshade = True)
    if show:
        plt.show()
    
    return ax
```

```python
ax = plt3D(phi(X))
```

```python
phi(X)
```

# Check Accuracy of X and X_ to Y     (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

#### Logistic Classifer in Higher Dimension Data

```python
lr3d = LogisticRegression()
acc = cross_val_score(lr3d, phi(X), Y, cv=5).mean()
print("Accuracy X(3D) is %.4f"%(acc*100))
```

#### Logistic Classifer on original  Data

```python
lr2d = LogisticRegression()
acc = cross_val_score(lr2d, X, Y, cv=5).mean()
print("Accuracy X(2D) is %.4f"%(acc*100))
```

# Add hyperplane to plot

```python
lr = LogisticRegression()
```

```python
lr.fit(phi(X),Y)
```

```python
wts = lr.coef_
print(wts)
```

```python
bias = lr.intercept_
print(bias)
```

```python
xx, yy = np.meshgrid(range(-1,2), range(-1,2))

print('xx: ',str(xx),'yy: ',str(yy), sep = '\n')

zz = -(wts[0][0]*xx+wts[0][1]*yy+bias)/wts[0][2]
print('zz :', str(zz),sep = '\n')
```

```python
ax = plt3D(phi(X),show = False)
ax.plot_surface(xx,yy,zz,alpha = 0.5)
plt.show()
```
