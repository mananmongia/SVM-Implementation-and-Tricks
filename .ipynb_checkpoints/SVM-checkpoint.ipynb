# Make Dataset

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np
```

```python
X,Y = make_classification(n_classes=2,n_samples=400,n_clusters_per_class=1,random_state=3,n_features=2,n_informative=2,n_redundant=0)
Y[Y == 0] = -1
```

```python
plt.scatter(X[:,0],X[:,1],c =Y)
plt.show()
```

# SVM Class Implementation

```python
class SVM:
    """Author: Manan Mongia"""
    def __init__(self,C = 1.0):
        self.C = C
        self.W = 0
        self.b = 0
        
    def hingeloss(self,C,W,b,X,Y):
        loss = 0.0
        
        loss = 0.5 * np.dot(W,W.T)
        
        m = X.shape[0]
        
        for i in range(m):
            ti = Y[i] * (np.dot(W.T,X[i]) + b)
            loss += self.C * max(0,1-ti)
            
        return loss
    
    def fit(self, X, Y,batch_size = 100, lr = 0.0001,maxItr=300):
        n = X.shape[1]
        m = X.shape[0]
        
        C = self.C
        W = np.zeros(n)
        bias = 0
        
        print("Intial Loss: ", str(self.hingeloss(C,W,bias,X,Y)))
        
        losses = []
        
        for i in range(maxItr):
            l = self.hingeloss(C,W,bias,X,Y)
            losses.append(l)
            #Batch Gradient Descent (Random Shuffling)
            ids = np.arange(m)
            np.random.shuffle(ids)
            
            for batch_start in range(0,m,batch_size):
                #Assume 0 gardient for the batch
                gradw = 0
                gradb = 0
                
                for j in range(batch_start,batch_start+batch_size):
                    if j<m :
                        i = ids[j]
                        ti =  Y[i]*(np.dot(W,X[i].T)+bias)
                        
                        if ti > 1 :
                            gradw += 0
                            gradb += 0
                        else:
                            gradw += C * Y[i] * X[i]
                            gradb += C * Y[i]
                
                W = W - lr*W + lr*gradw
                bias = bias + lr*gradb
            
                
        self.W = W
        self.b = bias
        
        print ("Final loss: "+ str(losses[-1]))
        print()
        print("W: " + str(self.W))
        print("Bias: " + str(self.b))
        
        return W,bias,losses
```

# Train

```python
mysvm = SVM()
```

```python
W,b,losses = mysvm.fit(X,Y,maxItr=3000)
```

```python
def plothyperplane(w1,w2,b):
    plt.figure(figsize=(12,10))
    plt.scatter(X[:,0],X[:,1],c =Y) 
    
    x1 = np.linspace(-2,4,10) 
    
    x2 = -(w1*x1 + b)/w2    # WT + B = 0
    x_p = -(w1*x1 + b + 1)/w2    # WT + B = -1
    x_n = -(w1*x1 + b - 1)/w2    # WT + B = +1
    
    plt.plot(x1,x2, label = "WT + B = 0          Hyperplane")
    plt.plot(x1,x_p, label = "WT + B = +1    +ve Hyperplane")
    plt.plot(x1,x_n, label = "WT + B = -1    -ve Hyperplane")
    
    plt.legend()
    plt.show()
```

```python
plothyperplane(W[0],W[1],b)
```
